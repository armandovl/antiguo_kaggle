```python
#importamos las bibliotecas de cajon
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
datos=np.loadtxt("../input/datos2.txt") #este lo usamos para traer los datos
datosdf=pd.DataFrame(datos)#los transformamos en un dataframe
datosdf.columns = ["var1","var2","res"] #agregamos unos encabezados
datosdf.head(10)#visualizamos los 10 primeros datos
```

```python
#graficamos las dos X (var1, var2) y la Y (res) la ponemos de color diferente

#%matplotlib inline #solo si no esta en linea

import seaborn as sns #biblioteca para gráficos
ax=sns.scatterplot(x="var1", y="var2", data=datosdf, hue="res", palette="Set1")
```

```python
from sklearn.model_selection import train_test_split #traemos libreria para segmentar
from sklearn.linear_model import LogisticRegression #traemos parte de libreria para regresión logística
```

```python
#definimos las variables
X=np.array(datosdf.drop(["res"],1))
y=np.array(datosdf["res"])
```

```python
#separamos los datos de entrenamiento y prueba 80-20
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2) #,random_state=777
print("son {} datos para entrenamiento y {} datos para prueba".format(X_train.shape[0],X_test.shape[0]))
```

```python
#hacemos el entrenamiento de regresión logística
algoritmoRL=LogisticRegression()
algoritmoRL.fit(X_train,y_train)
Y_predRL=algoritmoRL.predict(X_test)
print("Precisión Regresión Logística\nCon entrenamiento:{}".format(algoritmoRL.score(X_train,y_train)))
```

```python
d = {'test': y_test, 'predicho': Y_predRL}
nuevoDF=pd.DataFrame(d)
print(nuevoDF.head(10))
print(len(nuevoDF))
```

```python
def coef_confianza(test,predicho):
    numAciertos=0
    m=len(nuevoDF)
    for i in range(m):
        if test[i]==predicho[i]:
            numAciertos+=1
    return(100*numAciertos/m)
```

```python
print(coef_confianza(nuevoDF["test"],nuevoDF["predicho"]))
```

```python
algoritmoRL.score(X_test,y_test)*100
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
print("aciertos en subconjunto= " ,accuracy_score(y_test,Y_predRL))
```

```python
print(confusion_matrix(y_test,Y_predRL))
```

```python
print(pd.crosstab(y_test,Y_predRL))
```

```python
#predicción de clasificación mediante regresión logistica
pre=np.array([[5,2],[7,6],[3,6],[4,7],[7,0],[9,6]])
predicciones=algoritmoRL.predict(pre)
print(predicciones)
```

```python
#extra Predicción mediante Máquinas de vectores de soporte

#cargamos la parte de la biblioteca
from sklearn.svm import SVC

algoritmoSVM=SVC()
algoritmoSVM.fit(X_train,y_train)
Y_predSVM=algoritmoSVM.predict(X_test)
print("Precisión con máquina de vectores de soporte:{}".format(algoritmoSVM.score(X_train,y_train)))
```

```python
print(algoritmoSVM.score(X_test,y_test))
```

```python
print(pd.crosstab(y_test,Y_predSVM))
```

```python
#predicción con máquinas de vectores de soporte
pre=np.array([[5,2],[7,6],[3,6],[4,7],[7,0],[9,6]])
predicciones=algoritmoSVM.predict(pre)
print(predicciones)
```

```python
print(algoritmoVecinos.score(X_test,y_test))
```

```python
print(pd.crosstab(y_test,Y_predVecinos))
```
