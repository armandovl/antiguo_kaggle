```python
#importamos las bibliotecas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
print(tf.__version__) #imprime la versión de tensorflow que usamos
```

```python
datos=np.loadtxt("../input/datos2D2C.txt") #este lo usamos para traer los datos
datosdf=pd.DataFrame(datos)#los transformamos en un dataframe
datosdf.columns = ["var1","var2","res"] #agregamos unos encabezados
datosdf.head(20)#visualizamos los datos
```

```python
from sklearn.model_selection import train_test_split #traemos libreria para segmentar
from sklearn.linear_model import LogisticRegression #traemos parte de libreria para regresión logística
```

```python
#definimos las variables
X=np.array(datosdf.drop(["res"],1))
y=np.array(datosdf["res"])
```

```python
#serparamos los datos de entrenamiento y prueba
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
print("son {} datos para entrenamiento y {} datos para prueba".format(X_train.shape[0],X_test.shape[0]))
```

```python
#hacemos el entrenamiento de regresión logística
algoritmo=LogisticRegression()
algoritmo.fit(X_train,y_train)
Y_pred=algoritmo.predict(X_test)
print("Precisión de la Regresión Logística:{}".format(algoritmo.score(X_train,y_train)))
```

```python
pre=np.array([[5,2],[7,6],[3,6],[4,7],[7,0],[9,6]])
predicciones=algoritmo.predict(pre)
print(predicciones)
```
