```python
import numpy as np
from scipy.cluster import vq
import matplotlib.pyplot as plt
```

```python
datos=np.loadtxt("../input/datos2D3Clusters.txt")
print(datos[0:4,:])
```

```python
#graficando los datos
plt.scatter(datos[:,0],datos[:,1])
plt.show()
```

```python
#calcular los centroides usando kmeans
centroides,varianza=vq.kmeans(datos,3)
print(centroides)
print(varianza) #mide el error el -1 es un estimador insesgado y de minima varianza
```

```python
#calculando su grupo y las distancias
grupo, distancia= vq.vq(datos,centroides)
print(grupo)
print(distancia) #elevo las distancias al cuadrado y luego las sumo y las divido entre m
```

```python
#separando por grupos
grupo0=datos[grupo==0]
grupo1=datos[grupo==1]
grupo2=datos[grupo==2]
print(len(grupo0))
print(len(grupo1))
print(len(grupo2))
```

```python
#graficando los grupos
plt.scatter(grupo0[:,0],grupo0[:,1],color="red")
plt.scatter(grupo1[:,0],grupo1[:,1],color="green")
plt.scatter(grupo2[:,0],grupo2[:,1],color="yellow")
plt.plot(centroides[:,0],centroides[:,1],"kx",markersize=15)
plt.show()
```

```python
#pronosticando nuevos valores
datosNuevos=[[5,2],[7,6],[3,6],[4,7],[7,0],[9,6]]
grupo_pronosticado, distancia_pronosticada=vq.vq(datosNuevos,centroides)
print(grupo_pronosticado)
```
