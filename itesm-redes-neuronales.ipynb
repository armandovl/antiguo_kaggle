```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
print(tf.__version__) #imprime la versión de tensorflow que usamos
```

```python
#solo una parte de tf
from tensorflow.keras import models
from tensorflow.keras import layers
from tensorflow.keras.utils import to_categorical #convertir en binarios como hacer dummies
```

```python
datos=np.loadtxt("../input/datos2.txt") #este lo usamos para traer los datos
print(datos.shape)

#separarlos en X y Y
x= datos[:,:2]#la dos no la incluye
y= datos[:,2]
```

```python
#separarlos
m=len(y)
limite=int(0.8*m)
xTrain,xTest=x[:limite,:],x[limite:,:]
yTrain,yTest=y[:limite],y[limite:]
print(xTrain.shape,xTest.shape,yTrain.shape,yTest.shape)
```

```python
#definir su arquitectura de la red
network=models.Sequential() #secuencial es que reciben la de la capa anterior, otras redes que no reciben de la capa anterior son las convolutivas y recurrentes
network.add(layers.Dense(20,activation='relu',input_shape=(2,)))#pegar la capa intermedia denso es que todas las neuronas de una capa reciben las salidas de la capa anterior
network.add(layers.Dense(3,activation='softmax')) #sofmax es para la capa de salida
```

```python
#paso de compilacion
network.compile(optimizer='rmsprop',
                loss='categorical_crossentropy',
               metrics=['accuracy'])
```

```python
#preparando las etiquetas
print(yTrain[:10]) #imprimimos los primeros 10
yTrain=to_categorical(yTrain)
print(yTrain[:10]) #imprimimos los primeros 10

yTest=to_categorical(yTest)
```

```python
#entrenar la red
network.fit(xTrain,yTrain,epochs=25,batch_size=128)
```

**Se pueden subri las epocas y volverlo a correr**

```python
#falta probarla en test

test_loss,tess_acc=network.evaluate(xTest,yTest)
print(test_loss,tess_acc)
```

```python
#probando algunos valores
#debe dar [0 1 2 2 0 1]

pre=np.array([[5,2],[7,6],[3,6],[4,7],[7,0],[9,6]])
predicciones=network.predict(pre)
print(predicciones)
#usamos argmax
res=np.argmax(predicciones,axis=1) #el uno en el axis lo hace por filas
print(res)
```
