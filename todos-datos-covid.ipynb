```r
Datos1<-read.csv("../input/rrrrrr/arbol_covid222.csv")
head(Datos1,10)
```

```r
Datos1$CUIDADOS_INTENSIVOS<-NULL
```

```r
Datos1$INTUBADO<-NULL
```

```r
Datos1$MUNICIPIO<-NULL
```

```r
str(Datos1)
```

```r
library(rpart)
library(rpart.plot)

#modelo rpart
modelo_rpart<-rpart(ESTIMADOS_SINTOMAS~.,
                    method = "class",
                    data=Datos1)

#importancia del modelo
library(caret) # libreria de la importancia
dt_importance <- varImp(modelo_rpart)
print(dt_importance) #variables y su importancia en el modelo

summary(modelo_rpart) #características del modelo
```

```r
rpart.plot(modelo_rpart,tweak=0.8,extra=101,type=4)#,type=2,extra=100,tweak=0.8,compress=FALSE,cex=1.2
```

```r
rpart.plot(modelo_rpart,tweak=0.8,extra=106,type=4)#,type=2,extra=100,tweak=0.8,compress=FALSE,cex=1.2
```
