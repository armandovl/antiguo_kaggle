```python
#importamos las bibliotecas de cajon
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset_train=pd.read_csv("../input/2014.csv")
training_set=dataset_train.iloc[:,1:2].values
```

```python
#Normalizaci√≥n
from sklearn.preprocessing import MinMaxScaler
sc=MinMaxScaler(feature_range=(0,1))
training_set_escalado=sc.fit_transform(training_set)
```

```python
#estructura 60 pasos temporales y una estructutra
X_train=[]
y_train=[]

for i in range(60,2189):
    X_train.append(training_set_escalado[i-60:i,0])
    y_train.append(training_set_escalado[i,0])
X_train,y_train=np.array(X_train),np.array(y_train)
```

```python
#agregar dimensionalidad
X_train=np.reshape(X_train,(X_train.shape[0],X_train.shape[1],1))
```

```python
import tensorflow as tf
import keras
```

```python
from keras.models import Sequential
from keras.layers import Dense,LSTM,Dropout
```

```python
#iniciando el modelo
regresor=Sequential()
```

```python
regresor.add(LSTM(units=50,return_sequences=True,input_shape=(X_train.shape[0],X_train.shape[1],1))
regresor.add(Dropout(0.2))
```
